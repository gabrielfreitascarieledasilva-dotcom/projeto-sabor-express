```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


pedidos = np.array([
    [2, 3],  [3, 4],  [2, 5],
    [8, 1],  [9, 2],  [7, 3],
    [1, 8],  [2, 9],  [3, 7],  [2, 8]
])

numero_de_grupos = 3

kmeans = KMeans(n_clusters=numero_de_grupos, n_init='auto', random_state=42)

kmeans.fit(pedidos)

grupos_atribuidos = kmeans.labels_
centros_dos_grupos = kmeans.cluster_centers_

print("--- Resultados do Agrupamento K-Means ---")
print(f"Dividimos {len(pedidos)} pedidos em {numero_de_grupos} zonas.\n")

for i in range(numero_de_grupos):
    print(f"ðŸšš ZONA {i} (Entregador {i+1}):")

    pedidos_da_zona = np.where(grupos_atribuidos == i)[0]
    for pedido_indice in pedidos_da_zona:
        print(f"  - Pedido {pedido_indice + 1} (Coordenadas: {pedidos[pedido_indice]})")
    print(f"  -> Centro da Zona: {np.round(centros_dos_grupos[i], 2)}\n")

plt.figure(figsize=(8, 6))
plt.scatter(pedidos[:, 0], pedidos[:, 1], c=grupos_atribuidos, cmap='viridis', marker='o', s=100, label='Pedidos')
plt.scatter(centros_dos_grupos[:, 0], centros_dos_grupos[:, 1], c='red', marker='X', s=200, label='Centro da Zona (Entregador)')

plt.title('Agrupamento de Pedidos "Sabor Express" com K-Means')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G = nx.Graph()

pos = {
    'Restaurante': (0, 0),
    'Esquina_A': (2, 2),
    'Esquina_B': (5, 4),
    'Esquina_C': (3, 6),
    'Pedido_1': (1, 4),
    'Pedido_2': (4, 7),
    'Pedido_3': (7, 5)
}
G.add_nodes_from(pos.keys())

G.add_edge('Restaurante', 'Esquina_A', peso=3)
G.add_edge('Restaurante', 'Pedido_1', peso=5)
G.add_edge('Esquina_A', 'Pedido_1', peso=2)
G.add_edge('Esquina_A', 'Esquina_B', peso=4)
G.add_edge('Pedido_1', 'Esquina_C', peso=3)
G.add_edge('Esquina_B', 'Pedido_3', peso=2)
G.add_edge('Esquina_B', 'Pedido_2', peso=4)
G.add_edge('Esquina_C', 'Pedido_2', peso=2)

print("--- Mapa (Grafo) da Sabor Express Criado ---")
print(f"NÃ³s (Locais): {G.nodes()}")
print(f"Arestas (Ruas): {G.edges(data=True)}\n")

def heuristica_distancia(a, b):
    (x1, y1) = pos[a]
    (x2, y2) = pos[b]
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)


ponto_partida = 'Restaurante'
pedidos_da_zona = ['Pedido_1', 'Pedido_2', 'Pedido_3']

rota_completa = [ponto_partida]
custo_total = 0

print("--- Calculando a Rota Otimizada com A* ---")

for proximo_pedido in pedidos_da_zona:
    caminho = nx.astar_path(G, ponto_partida, proximo_pedido,
                              heuristic=heuristica_distancia, weight='peso')


    custo_trecho = nx.astar_path_length(G, ponto_partida, proximo_pedido,
                                          heuristic=heuristica_distancia, weight='peso')

    print(f"Trecho: {ponto_partida} -> {proximo_pedido}")
    print(f"  Melhor caminho: {caminho[1:]}")
    print(f"  Custo do trecho: {custo_trecho}")


    rota_completa.extend(caminho[1:])
    custo_total += custo_trecho


    ponto_partida = proximo_pedido

print("\n--- ROTA FINAL OTIMIZADA ---")
print(f"SequÃªncia de visita: {' -> '.join(rota_completa)}")
print(f"Custo total da rota: {custo_total}")




plt.figure(figsize=(10, 7))

nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold')

labels_peso = nx.get_edge_attributes(G, 'peso')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels_peso)

edges_da_rota = list(zip(rota_completa, rota_completa[1:]))
nx.draw_networkx_nodes(G, pos, nodelist=rota_completa, node_color='lightgreen', node_size=2000)
nx.draw_networkx_edges(G, pos, edgelist=edges_da_rota, edge_color='red', width=3)


plt.title('Mapa da "Sabor Express" com Rota Otimizada ($A^{*}$)')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans


pos = {
    'Restaurante': (5, 0),

    'E_A': (3, 2), 'E_B': (7, 2), 'E_C': (3, 6), 'E_D': (7, 7),

    'P1': (2, 3), 'P2': (3, 4), 'P3': (2, 5),
    'P4': (8, 1), 'P5': (9, 2), 'P6': (7, 3),
    'P7': (1, 8), 'P8': (2, 9), 'P9': (3, 7), 'P10': (2, 8)
}


pedidos_nomes = [f'P{i+1}' for i in range(10)]
pedidos_coords = np.array([pos[nome] for nome in pedidos_nomes])


G = nx.Graph()
G.add_nodes_from(pos.keys())


def heuristica_distancia(a, b):
    (x1, y1) = pos[a]
    (x2, y2) = pos[b]
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)


G.add_edge('Restaurante', 'E_A', peso=heuristica_distancia('Restaurante', 'E_A'))
G.add_edge('Restaurante', 'E_B', peso=heuristica_distancia('Restaurante', 'E_B'))
G.add_edge('E_A', 'P1', peso=heuristica_distancia('E_A', 'P1'))
G.add_edge('E_A', 'P2', peso=heuristica_distancia('E_A', 'P2'))
G.add_edge('E_B', 'P4', peso=heuristica_distancia('E_B', 'P4'))
G.add_edge('E_B', 'P5', peso=heuristica_distancia('E_B', 'P5'))
G.add_edge('E_B', 'P6', peso=heuristica_distancia('E_B', 'P6'))
G.add_edge('P1', 'P3', peso=heuristica_distancia('P1', 'P3'))
G.add_edge('P2', 'P3', peso=heuristica_distancia('P2', 'P3'))
G.add_edge('P2', 'E_C', peso=heuristica_distancia('P2', 'E_C'))
G.add_edge('P3', 'E_C', peso=heuristica_distancia('P3', 'E_C'))
G.add_edge('E_C', 'P9', peso=heuristica_distancia('E_C', 'P9'))
G.add_edge('E_C', 'P10', peso=heuristica_distancia('E_C', 'P10'))
G.add_edge('E_D', 'P6', peso=heuristica_distancia('E_D', 'P6'))
G.add_edge('E_D', 'P8', peso=heuristica_distancia('E_D', 'P8'))
G.add_edge('P7', 'P8', peso=heuristica_distancia('P7', 'P8'))
G.add_edge('P7', 'P10', peso=heuristica_distancia('P7', 'P10'))
G.add_edge('P8', 'P10', peso=heuristica_distancia('P8', 'P10'))
G.add_edge('P9', 'P10', peso=heuristica_distancia('P9', 'P10'))


print("--- MAPA (GRAFO) CRIADO ---")



numero_de_grupos = 3
kmeans = KMeans(n_clusters=numero_de_grupos, n_init='auto', random_state=42)
kmeans.fit(pedidos_coords)
grupos_atribuidos = kmeans.labels_

print("\n--- ZONAS DEFINIDAS (K-MEANS) ---")
print(f"Pedidos: {pedidos_nomes}")
print(f"Grupos:  {grupos_atribuidos}")



print("\n--- CALCULANDO ROTAS OTIMIZADAS (A*) ---")


plt.figure(figsize=(12, 10))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000, font_size=9)
labels_peso = nx.get_edge_attributes(G, 'peso')

labels_peso_arredondado = {k: f"{v:.1f}" for k, v in labels_peso.items()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels_peso_arredondado, font_size=7)


cores_rotas = ['red', 'blue', 'green']
all_rotas_completas = []


for i in range(numero_de_grupos):


    indices_zona = np.where(grupos_atribuidos == i)[0]


    pedidos_da_zona = [pedidos_nomes[idx] for idx in indices_zona]

    print(f"\nðŸšš ROTA PARA ENTREGADOR {i+1} (Cor: {cores_rotas[i].upper()})")
    print(f"   Pedidos: {pedidos_da_zona}")

    if not pedidos_da_zona:
        print("   Nenhum pedido nesta zona.")
        continue


    ponto_partida = 'Restaurante'
    rota_completa_zona = [ponto_partida]
    custo_total_zona = 0


    pedidos_restantes = list(pedidos_da_zona)

    while pedidos_restantes:

        melhor_pedido = None
        menor_custo = float('inf')

        for pedido in pedidos_restantes:
            try:
                custo = nx.astar_path_length(G, ponto_partida, pedido,
                                            heuristic=heuristica_distancia, weight='peso')
                if custo < menor_custo:
                    menor_custo = custo
                    melhor_pedido = pedido
            except nx.NetworkXNoPath:

                print(f"AVISO: NÃ£o hÃ¡ caminho de {ponto_partida} para {pedido}")
                continue

        if melhor_pedido is None:
            print(f"   NÃ£o foi possÃ­vel encontrar rota para {pedidos_restantes}")
            break


        caminho_trecho = nx.astar_path(G, ponto_partida, melhor_pedido,
                                        heuristic=heuristica_distancia, weight='peso')

        print(f"  -> Trecho: {ponto_partida} -> {melhor_pedido} (Custo: {menor_custo:.2f})")

        rota_completa_zona.extend(caminho_trecho[1:])
        custo_total_zona += menor_custo


        ponto_partida = melhor_pedido
        pedidos_restantes.remove(melhor_pedido)

    print(f"   ROTA FINAL ({cores_rotas[i]}): {' -> '.join(rota_completa_zona)}")
    print(f"   CUSTO TOTAL: {custo_total_zona:.2f}")


    all_rotas_completas.append((rota_completa_zona, cores_rotas[i]))




print("\n--- DESENHANDO GRÃFICO FINAL DAS ROTAS ---")


for rota, cor in all_rotas_completas:
    edges_da_rota = list(zip(rota, rota[1:]))
    nx.draw_networkx_nodes(G, pos, nodelist=rota, node_color=cor, node_size=1000)
    nx.draw_networkx_edges(G, pos, edgelist=edges_da_rota, edge_color=cor, width=3)

plt.title('Rotas Otimizadas "Sabor Express" (K-Means + $A^{*}$)')
plt.show()
```
